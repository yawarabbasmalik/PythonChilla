# Line Plot with Multi-Facets

The Bar plot is mostly used for Plotting of both numeric and categorical data 

```python
#importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing data
clean_data = pd.read_csv("Chilla_data2_for_plots.csv")
clean_data.head()
```

```python
# Color Changing in Bar Plot
sns.relplot(x = "Age (years)-Float/Int", y = "What are you?",
            hue = "Gender", data = clean_data, palette="Accent_r",
            kind = "line",
            height = 5, aspect=0.75, facet_kws=dict(sharex=False),
           )
plt.show()
```

# Displot (Histogram)

```python
sns.displot(
    data = clean_data, x="Age (years)-Float/Int", col="What are you?", row="Gender",
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
```

# Grouped barplotsÂ¶

```python
g = sns.catplot(
    data = clean_data, x="Age (years)-Float/Int", y="What are you?", hue="Gender",
    kind="bar", ci=None, palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Profession")
g.legend.set_title("Python Chilla Students Data")
```

# Joint Plot

```python
g = sns.jointplot(
    data=clean_data,
    x="Age (years)-Float/Int", y="How many hours you code a day? (int) e.g: 5,4,3", hue="Gender",
    kind="kde",
)
```
