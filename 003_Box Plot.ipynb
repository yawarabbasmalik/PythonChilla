# Box Plot

The Bar plot is mostly used for Plotting of both numeric and categorical data 

```python
#importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing data
clean_data = pd.read_csv("Chilla_data2_for_plots.csv")
clean_data.head()
```

```python
#drawing a boxplot
sns.boxplot(x = "Your favorite programming language?", 
            y = "How many hours you code a day? (int) e.g: 5,4,3", 
            data = clean_data)
plt.show()
```

# Box Plot Style

```python
sns.set(style='whitegrid')
#drawing a boxplot
sns.boxplot(x = "Your favorite programming language?", 
            y = "How many hours you code a day? (int) e.g: 5,4,3", 
            data = clean_data)
plt.show()
```

# Categoriazation in Bar Plot

```python
sns.set(style='whitegrid')
#drawing a boxplot
sns.boxplot(x = "Your favorite programming language?", 
            y = "How many hours you code a day? (int) e.g: 5,4,3", 
            data = clean_data, palette="Accent_r",
           saturation=2)
plt.show()
```

# Important Notes
- Categorical Data (Attributes) are mostly drawn on x-axis.
- Numeric Data (Attributes) are mostly drawn on y-axis.
- Numeric Data (Attributes) are never drwan using "Hue".
- The box contains the value of interquartile range.
- The center bar in the box represents "Median".
- The Right/ Left or sometime Upper/Lower bars represents the maximum and minimum outlier data.
- The dots represents the outliers

```python
sns.set(style='whitegrid')
#drawing a boxplot
sns.boxplot(x = clean_data["Age (years)-Float/Int"])
plt.show()
```

# Color Picking

```python
sns.set(style='whitegrid')
#drawing a boxplot
sns.boxplot(x = "Your favorite programming language?", 
            y = "How many hours you code a day? (int) e.g: 5,4,3", 
            data = clean_data, color="#09ebdc")
plt.show()
```

# Showing Means in the Box Plot

```python
sns.set(style='whitegrid')
#drawing a boxplot
sns.boxplot(x = "Your favorite programming language?", 
            y = "How many hours you code a day? (int) e.g: 5,4,3",
            showmeans = True,
            data = clean_data, color="#09ebdc")
plt.show()
```

# Setting Mean Properties in Box Plot

```python
sns.set(style='whitegrid')
#drawing a boxplot
sns.boxplot(x = "Your favorite programming language?", 
            y = "How many hours you code a day? (int) e.g: 5,4,3",
            showmeans = True,
            meanprops = { "marker" : "+",
                         "markersize" : "10",
                         "markeredgecolor" : "#464694"},
            data = clean_data, color="#09ebdc")
plt.show()
```

# Adusting Sizes and Labels

```python
sns.set(style='whitegrid')
#drawing a boxplot
sns.boxplot(x = "Your favorite programming language?", 
            y = "How many hours you code a day? (int) e.g: 5,4,3",
            showmeans = True,
            meanprops = { "marker" : "+",
                         "markersize" : "10",
                         "markeredgecolor" : "#464694"},
            data = clean_data, color="#09ebdc")

plt.xlabel("Programmers Lanaguges", size = 12)
plt.ylabel("Programmers Practice Each Day", size = 12)
plt.title("Chilla Programmers Stats", size = 18, weight = "bold")
plt.show()
```
