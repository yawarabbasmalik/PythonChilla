# Line Plot

The Line plot is mostly used for Plotting of numeric values rather than categorical values 

```python
#importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing data
clean_data = pd.read_csv("Chilla_data2_for_plots.csv")
clean_data.head()
```

```python
#drawing a lineplot
sns.lineplot(x = "Age (years)-Float/Int", y = "Height in cm? Freelancer- (Float)", data = clean_data)
plt.show()
```

# Adding Title

```python
#drawing a lineplot
sns.lineplot(x = "Age (years)-Float/Int", y = "Height in cm? Freelancer- (Float)", data = clean_data)
plt.title("Python Chilla Class Age & Height Stats")
plt.show()
```

# Adding Limits x&y

```python
#Adding limits
sns.lineplot(x = "Age (years)-Float/Int", y = "Height in cm? Freelancer- (Float)", data = clean_data)
plt.title("Python Chilla Class Age & Height Stats")
plt.xlim(5)
plt.ylim(100)
plt.show()
```

# Style Setting

```python
sns.set_style(style=None, rc=None)
```

```python
# Setting Style
sns.lineplot(x = "Age (years)-Float/Int", y = "Height in cm? Freelancer- (Float)", data = clean_data)
plt.title("Python Chilla Class Age & Height Stats")
sns.set_style("dark")
plt.show()
```

# Size Adjustment

```python
#Setting size of the Plot
plt.figure(figsize=(8, 6))

sns.lineplot(x = "Age (years)-Float/Int", y = "Height in cm? Freelancer- (Float)", data = clean_data)
plt.title("Python Chilla Class Age & Height Stats")
plt.show()
```
