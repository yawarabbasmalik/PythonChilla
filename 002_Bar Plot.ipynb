# Bar Plot

The Bar plot is mostly used for Plotting of both numeric and categorical data 

```python
#importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing data
clean_data = pd.read_csv("Chilla_data2_for_plots.csv")
clean_data.head()
```

```python
#drawing a lineplot
sns.barplot(x = "Your favorite programming language?", y = "How many hours you code a day? (int) e.g: 5,4,3", data = clean_data)
plt.show()
```

# Categoriazation in Bar Plot

```python
# Categoriazation in Bar Plot
sns.barplot(x = "Your favorite programming language?", y = "How many hours you code a day? (int) e.g: 5,4,3",
            hue = "Gender", data = clean_data)
plt.show()
```

# Color Changing in Bar Plot

```python
# Color Changing in Bar Plot
sns.barplot(x = "Your favorite programming language?", y = "How many hours you code a day? (int) e.g: 5,4,3",
            hue = "Gender", data = clean_data, color="red")
plt.show()
```

# Removing Confidence Interval

```python
# Color Changing in Bar Plot
sns.barplot(x = "Your favorite programming language?", y = "How many hours you code a day? (int) e.g: 5,4,3",
            hue = "Gender", data = clean_data, color="red", ci = None)
plt.show()
```

# Adding Color Palettes

```python
# Color Changing in Bar Plot
sns.barplot(x = "Your favorite programming language?", y = "How many hours you code a day? (int) e.g: 5,4,3",
            hue = "Gender", data = clean_data, color="red", ci = None, palette="Accent_r")
plt.show()
```

# Adjusitng Color's Saturation

```python
# Color Saturation Adjustment
sns.barplot(x = "Your favorite programming language?", y = "How many hours you code a day? (int) e.g: 5,4,3",
            hue = "Gender", data = clean_data, color="red", ci = None, palette="Accent_r",
           saturation=2)
plt.show()
```

# Plotting a Horizontal Plot

```python
# Plotting a Horizontal Bar Graph
sns.barplot(x = "How many hours you code a day? (int) e.g: 5,4,3", y = "Your favorite programming language?",
            hue = "Gender", data = clean_data, color="red", ci = None, palette="Accent_r",
           saturation=0.2)
plt.show()
```

# Using Custom Values Rather than "Hue"

```python
# Plotting a Horizontal Bar Graph
sns.barplot(x = "Your favorite programming language?", y = "How many hours you code a day? (int) e.g: 5,4,3",
            data = clean_data,
            linewidth=2.5, facecolor = (1,1,1,1),
            errcolor="0.2", edgecolor= "0.5")

plt.show()
```
